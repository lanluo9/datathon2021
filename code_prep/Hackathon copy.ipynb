```python
#A function that drops all columns containing NaN values except for the column clmn from which all NaN rows are removed

def remove_all_nan_cols_but_one(clmn, df_in):
    cols_to_drop = []
    for c in df_in.columns:
        if df_in[c].isnull().values.any():
            cols_to_drop += [c]
    cols_to_drop.remove(clmn)
    df_out = df_in.copy().drop(cols_to_drop, axis=1, inplace=False)
    if df_out[clmn].isnull().values.any():
        df_out = df_out.loc[~df_out[clmn].isnull()]
    return df_out

# # define baseline model
# def baseline_model(num_inputs=None, num_outputs=None):
#     # create model
#     model = Sequential()
#     model.add(Dense((num_inputs+num_outputs)/2, input_dim=num_inputs, activation='relu'))
#     model.add(Dense(num_outputs, activation='softmax'))
#     # Compile model
#     model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
#     return model
```

```python
# function to print sentiments
# of the sentence.
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        return "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
        return "Negative"
 
    else :
        return "Neutral"
```
